Write a function make_deck that returns a list of all of the cards in a standard card deck. The return should be a list of tuples of pairs of suit and value. For example the 10 of Clubs would be ('Clubs', 10) and Queen of Hearts would be ('Hearts', 'Queen'). Recall that a deck has 52 cards, divided into 4 suits (Clubs, Diamonds, Hearts, and Spades), and that each suit has 13 cards: 2 to 10, Jack, Queen, King, and Ace. Summit your solution with Lab 5.

```python
def make_deck():
    # Define the card suits
    suits = ['Clubs', 'Diamonds', 'Hearts', 'Spades']

    # Define the card values
    values = [2, 3, 4, 5, 6, 7, 8, 9, 10, 'Jack', 'Queen', 'King', 'Ace']

    # Create the deck using list comprehension
    deck = [(suit, value) for suit in suits for value in values]

    return deck

# Execute the function and check the result
deck = make_deck()
print(deck)
```

1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class Matrix:
    def __init__(self, n, m=None):
        """
        Initialize a matrix in two ways:
        1. If two integers (n, m) are provided, create an n x m matrix filled with zeros.
        2. If a list of lists is provided, initialize the matrix with given values.
        """
        if isinstance(n, int) and isinstance(m, int):  # Case 1: Initialize with size
            self.data = [[0] * m for _ in range(n)]
        elif isinstance(n, list):  # Case 2: Initialize with list of lists
            self.data = n
            row_lengths = {len(row) for row in n}  # Ensure all rows have the same number of columns
            if len(row_lengths) > 1:
                raise ValueError("All rows must have the same number of columns.")
        else:
            raise TypeError("Invalid initialization parameters.")

    def __repr__(self):
        """
        Return a string representation of the matrix.
        """
        return "\n".join([" ".join(map(str, row)) for row in self.data])

# Example usage
M1 = Matrix(3, 3)  # Create a 3x3 matrix filled with zeros
print("Matrix initialized with size 3x3:")
print(M1)

M2 = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])  # Initialize with a list of lists
print("\nMatrix initialized with a list of lists:")
print(M2)
```

2. Add the following methods:
    * `shape()`: returns a tuple `(n,m)` of the shape of the matrix.
    * `transpose()`: returns a new matrix instance which is the transpose of the matrix.
    * `row(n)` and `column(n)`: that return the nth row or column of the matrix M as a new appropriately shaped matrix object.
    * `to_list()`: which returns the matrix as a list of lists.
    *  `block(n_0,n_1,m_0,m_1)` that returns a smaller matrix located at the n_0 to n_1 columns and m_0 to m_1 rows.
    * (Extra credit) Modify `__getitem__` implemented above to support slicing.
        

```python
class Matrix:
    def __init__(self, n, m=None):
        """
        Initialize a matrix in two ways:
        1. If two integers (n, m) are provided, create an n x m matrix filled with zeros.
        2. If a list of lists is provided, initialize the matrix with given values.
        """
        if isinstance(n, int) and isinstance(m, int):  # Case 1: Initialize with size
            self.data = [[0] * m for _ in range(n)]
        elif isinstance(n, list):  # Case 2: Initialize with list of lists
            self.data = n
            row_lengths = {len(row) for row in n}  # Ensure all rows have the same number of columns
            if len(row_lengths) > 1:
                raise ValueError("All rows must have the same number of columns.")
        else:
            raise TypeError("Invalid initialization parameters.")

    def shape(self):
        """
        Returns a tuple (n, m) representing the shape of the matrix.
        """
        return len(self.data), len(self.data[0])

    def transpose(self):
        """
        Returns a new matrix instance which is the transpose of the matrix.
        """
        transposed_data = list(map(list, zip(*self.data)))  # Transpose using zip
        return Matrix(transposed_data)

    def row(self, n):
        """
        Returns the nth row of the matrix as a new Matrix object.
        """
        return Matrix([self.data[n]])

    def column(self, n):
        """
        Returns the nth column of the matrix as a new Matrix object.
        """
        col = [[row[n]] for row in self.data]  # Extract column as a list of lists
        return Matrix(col)

    def to_list(self):
        """
        Returns the matrix as a list of lists.
        """
        return self.data

    def block(self, n_0, n_1, m_0, m_1):
        """
        Returns a smaller matrix located at the given column (n_0 to n_1)
        and row (m_0 to m_1) range.
        """
        submatrix = [row[n_0:n_1] for row in self.data[m_0:m_1]]
        return Matrix(submatrix)

    def __getitem__(self, indices):
        """
        Modify __getitem__ to support slicing.
        Allows M[i][j] and M[i, j] for indexing, and supports slicing.
        """
        if isinstance(indices, tuple):
            if isinstance(indices[0], slice) or isinstance(indices[1], slice):
                # Extract rows based on slicing
                sliced_data = [row[indices[1]] for row in self.data[indices[0]]]
                return Matrix(sliced_data)
            i, j = indices
            return self.data[i][j]
        return self.data[indices]

    def __repr__(self):
        """
        Return a string representation of the matrix.
        """
        return "\n".join([" ".join(map(str, row)) for row in self.data])


# Example Usage
M = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])

print("Original Matrix:")
print(M)

print("\nMatrix Shape:")
print(M.shape())

print("\nTransposed Matrix:")
print(M.transpose())

print("\nRow 1:")
print(M.row(1))

print("\nColumn 2:")
print(M.column(2))

print("\nMatrix as list of lists:")
print(M.to_list())

print("\nSubmatrix from (0,2) columns and (1,3) rows:")
print(M.block(0, 2, 1, 3))

print("\nSlicing Example M[0:2, 1:3]:")
print(M[0:2, 1:3])
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
    """
    Returns an n by m matrix filled with floats of value c.
    """
    return Matrix([[float(c)] * m for _ in range(n)])

def zeros(n, m):
    """
    Returns an n by m matrix filled with floats of value 0.
    """
    return constant(n, m, 0.0)  # Uses the constant function

def ones(n, m):
    """
    Returns an n by m matrix filled with floats of value 1.
    """
    return constant(n, m, 1.0)  # Uses the constant function

def eye(n):
    """
    Returns the n by n identity matrix.
    """
    return Matrix([[1.0 if i == j else 0.0 for j in range(n)] for i in range(n)])

# Example usage
print("Constant Matrix (3x3) with value 5:")
print(constant(3, 3, 5))

print("\nZeros Matrix (2x4):")
print(zeros(2, 4))

print("\nOnes Matrix (3x3):")
print(ones(3, 3))

print("\nIdentity Matrix (4x4):")
print(eye(4))
```

4. Add the following member functions to your class. Make sure to appropriately test the dimensions of the matrices to make sure the operations are correct.
    * `M.scalarmul(c)`: a matrix that is scalar product $cM$, where every element of $M$ is multiplied by $c$.
    * `M.add(N)`: adds two matrices $M$ and $N$. Don’t forget to test that the sizes of the matrices are compatible for this and all other operations.
    * `M.sub(N)`: subtracts two matrices $M$ and $N$.
    * `M.mat_mult(N)`: returns a matrix that is the matrix product of two matrices $M$ and $N$.
    * `M.element_mult(N)`: returns a matrix that is the element-wise product of two matrices $M$ and $N$.
    * `M.equals(N)`: returns true/false if $M==N$.

```python
class Matrix:
    def __init__(self, n, m=None):
        """
        Initialize a matrix in two ways:
        1. If two integers (n, m) are provided, create an n x m matrix filled with zeros.
        2. If a list of lists is provided, initialize the matrix with given values.
        """
        if isinstance(n, int) and isinstance(m, int):  # Case 1: Initialize with size
            self.data = [[0] * m for _ in range(n)]
        elif isinstance(n, list):  # Case 2: Initialize with list of lists
            self.data = n
            row_lengths = {len(row) for row in n}  # Ensure all rows have the same number of columns
            if len(row_lengths) > 1:
                raise ValueError("All rows must have the same number of columns.")
        else:
            raise TypeError("Invalid initialization parameters.")

    def shape(self):
        """
        Returns a tuple (n, m) representing the shape of the matrix.
        """
        return len(self.data), len(self.data[0])

    def scalarmul(self, c):
        """
        Returns a new matrix that is the scalar product cM.
        Each element in the matrix is multiplied by c.
        """
        return Matrix([[c * value for value in row] for row in self.data])

    def add(self, N):
        """
        Returns a new matrix that is the sum of self (M) and N.
        Matrices must have the same dimensions.
        """
        if self.shape() != N.shape():
            raise ValueError("Matrix dimensions must match for addition.")
        return Matrix([[self.data[i][j] + N.data[i][j] for j in range(self.shape()[1])] for i in range(self.shape()[0])])

    def sub(self, N):
        """
        Returns a new matrix that is the difference of self (M) and N.
        Matrices must have the same dimensions.
        """
        if self.shape() != N.shape():
            raise ValueError("Matrix dimensions must match for subtraction.")
        return Matrix([[self.data[i][j] - N.data[i][j] for j in range(self.shape()[1])] for i in range(self.shape()[0])])

    def mat_mult(self, N):
        """
        Returns a new matrix that is the matrix product of self (M) and N.
        The number of columns in M must match the number of rows in N.
        """
        if self.shape()[1] != N.shape()[0]:
            raise ValueError("Number of columns in M must match number of rows in N for matrix multiplication.")

        result = [[sum(self.data[i][k] * N.data[k][j] for k in range(self.shape()[1])) for j in range(N.shape()[1])]
                  for i in range(self.shape()[0])]
        return Matrix(result)

    def element_mult(self, N):
        """
        Returns a new matrix that is the element-wise product of self (M) and N.
        Matrices must have the same dimensions.
        """
        if self.shape() != N.shape():
            raise ValueError("Matrix dimensions must match for element-wise multiplication.")
        return Matrix([[self.data[i][j] * N.data[i][j] for j in range(self.shape()[1])] for i in range(self.shape()[0])])

    def equals(self, N):
        """
        Returns True if self (M) and N are equal, False otherwise.
        """
        if self.shape() != N.shape():
            return False
        return all(self.data[i][j] == N.data[i][j] for i in range(self.shape()[0]) for j in range(self.shape()[1]))

    def __repr__(self):
        """
        Returns a string representation of the matrix.
        """
        return "\n".join([" ".join(map(str, row)) for row in self.data])


# Example Usage
M1 = Matrix([[1, 2], [3, 4]])
M2 = Matrix([[5, 6], [7, 8]])

print("Scalar Multiplication by 2:")
print(M1.scalarmul(2))

print("\nMatrix Addition (M1 + M2):")
print(M1.add(M2))

print("\nMatrix Subtraction (M1 - M2):")
print(M1.sub(M2))

M3 = Matrix([[1, 2, 3], [4, 5, 6]])
M4 = Matrix([[7, 8], [9, 10], [11, 12]])

print("\nMatrix Multiplication (M3 * M4):")
print(M3.mat_mult(M4))

print("\nElement-wise Multiplication (M1 * M2):")
print(M1.element_mult(M2))

print("\nEquality Check (M1 == M2):")
print(M1.equals(M2))

print("\nEquality Check (M1 == M1):")
print(M1.equals(M1))
```

5. Overload python operators to appropriately use your functions in 4 and allow expressions like:
    * 2*M
    * M*2
    * M+N
    * M-N
    * M*N
    * M==N
    * M=N


```python
class Matrix:
    def __init__(self, n, m=None):
        """
        Initialize a matrix in two ways:
        1. If two integers (n, m) are provided, create an n x m matrix filled with zeros.
        2. If a list of lists is provided, initialize the matrix with given values.
        """
        if isinstance(n, int) and isinstance(m, int):  # Case 1: Initialize with size
            self.data = [[0] * m for _ in range(n)]
        elif isinstance(n, list):  # Case 2: Initialize with list of lists
            self.data = n
            row_lengths = {len(row) for row in n}  # Ensure all rows have the same number of columns
            if len(row_lengths) > 1:
                raise ValueError("All rows must have the same number of columns.")
        else:
            raise TypeError("Invalid initialization parameters.")

    def shape(self):
        """
        Returns a tuple (n, m) representing the shape of the matrix.
        """
        return len(self.data), len(self.data[0])

    def __mul__(self, other):
        """
        Overloads the * operator.
        - If other is a scalar (int/float), performs scalar multiplication.
        - If other is a matrix, performs matrix multiplication.
        """
        if isinstance(other, (int, float)):  # Scalar multiplication
            return Matrix([[other * value for value in row] for row in self.data])
        elif isinstance(other, Matrix):  # Matrix multiplication
            if self.shape()[1] != other.shape()[0]:
                raise ValueError("Number of columns in M must match number of rows in N for matrix multiplication.")
            result = [[sum(self.data[i][k] * other.data[k][j] for k in range(self.shape()[1]))
                       for j in range(other.shape()[1])] for i in range(self.shape()[0])]
            return Matrix(result)
        else:
            raise TypeError("Multiplication is only supported for scalars or other matrices.")

    def __rmul__(self, other):
        """
        Allows scalar multiplication in the form: scalar * M (e.g., 2 * M).
        """
        return self * other  # Uses __mul__

    def __add__(self, other):
        """
        Overloads the + operator to add two matrices.
        """
        if not isinstance(other, Matrix) or self.shape() != other.shape():
            raise ValueError("Matrix dimensions must match for addition.")
        return Matrix([[self.data[i][j] + other.data[i][j] for j in range(self.shape()[1])] for i in range(self.shape()[0])])

    def __sub__(self, other):
        """
        Overloads the - operator to subtract two matrices.
        """
        if not isinstance(other, Matrix) or self.shape() != other.shape():
            raise ValueError("Matrix dimensions must match for subtraction.")
        return Matrix([[self.data[i][j] - other.data[i][j] for j in range(self.shape()[1])] for i in range(self.shape()[0])])

    def __eq__(self, other):
        """
        Overloads the == operator to compare two matrices.
        """
        if not isinstance(other, Matrix) or self.shape() != other.shape():
            return False
        return all(self.data[i][j] == other.data[i][j] for i in range(self.shape()[0]) for j in range(self.shape()[1]))

    def __repr__(self):
        """
        Returns a string representation of the matrix.
        """
        return "\n".join([" ".join(map(str, row)) for row in self.data])


# Example Usage
M1 = Matrix([[1, 2], [3, 4]])
M2 = Matrix([[5, 6], [7, 8]])

print("Scalar Multiplication (2 * M1):")
print(2 * M1)

print("\nScalar Multiplication (M1 * 2):")
print(M1 * 2)

print("\nMatrix Addition (M1 + M2):")
print(M1 + M2)

print("\nMatrix Subtraction (M1 - M2):")
print(M1 - M2)

M3 = Matrix([[1, 2, 3], [4, 5, 6]])
M4 = Matrix([[7, 8], [9, 10], [11, 12]])

print("\nMatrix Multiplication (M3 * M4):")
print(M3 * M4)

print("\nEquality Check (M1 == M2):")
print(M1 == M2)

print("\nEquality Check (M1 == M1):")
print(M1 == M1)
```
